```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_pickle('data.pkl')
df
```

```python
X_test = df[df['election_date'] == '2019-10-21'].iloc[:,:-1]
y_test = df[df['election_date'] == '2019-10-21'].iloc[:,-1]
X_train = df[~(df['election_date'] == '2019-10-21')].iloc[:,:-1]
y_train = df[~(df['election_date'] == '2019-10-21')].iloc[:,-1]
```

```python
import re
def preprocess_text(text):
    text = text.lower()
    text = text.replace('\n', ' ')
    text = re.sub(r'\d+', '', text)
    return text
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english', preprocessor = preprocess_text, min_df =0.10)
vectorizer.fit(X_train['text_en'])
X = vectorizer.transform(X_train['text_en'])
```

```python
vectorized = X.todense()
```

```python
vec_df = pd.DataFrame(vectorized, columns = vectorizer.vocabulary_)
```

```python
y_train = y_train.astype('int')
```

```python
from sklearn.linear_model import LogisticRegression
my_model = LogisticRegression(n_jobs=-1)
my_model.fit(vec_df, y_train)
```
